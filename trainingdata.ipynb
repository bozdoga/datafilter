```python
import pandas as pd 
path= 'C://Users/bozdoga/Documents/Data Filtering/dummydata.xlsx'
folder_path= 'C://Users/bozdoga/Documents/Data Filtering/output/'

```

```python
data = pd.read_excel(path)
data.describe()
```

```python
data_onlyde  = data[(data.active_marketplaces)=='DE']
data_allde = data[(data.active_marketplaces.str.contains('DE'))]
data_de = pd.concat([data_allde,data_onlyde]).drop_duplicates(keep=False)


data_de.head(200)
```

```python
data_onlyse  = data[(data.active_marketplaces)=='SE']
data_allse = data[(data.active_marketplaces.str.contains('SE'))]
data_se = pd.concat([data_allse,data_onlyde]).drop_duplicates(keep=False)

data_onlydtr  = data[(data.active_marketplaces)=='TR']
data_alltr = data[(data.active_marketplaces.str.contains('TR'))]
data_tr = pd.concat([data_alltr,data_onlyde]).drop_duplicates(keep=False)

data_onlydpl  = data[(data.active_marketplaces)=='PL']
data_allpl = data[(data.active_marketplaces.str.contains('PL'))]
data_pl = pd.concat([data_allpl,data_onlyde]).drop_duplicates(keep=False)
```

```python
data_se.head(100)
```

```python
with pd.ExcelWriter(folder_path +'DE Training Data.xlsx') as writer:
    data_onlyde.to_excel(writer,sheet_name='Only DE ASINs')
    data_de.to_excel(writer,sheet_name='DE ASINs')
```

```python
with pd.ExcelWriter(folder_path +'SE Training Data.xlsx') as writer:
    data_onlyse.to_excel(writer,sheet_name='Only SE ASINs')
    data_se.to_excel(writer,sheet_name='SE ASINs')

with pd.ExcelWriter(folder_path +'TR Training Data.xlsx') as writer:
    data_onlydtr.to_excel(writer,sheet_name='Only TR ASINs')
    data_tr.to_excel(writer,sheet_name='TR ASINs')

with pd.ExcelWriter(folder_path +'PL Training Data.xlsx') as writer:
    data_onlydpl.to_excel(writer,sheet_name='Only PL ASINs')
    data_pl.to_excel(writer,sheet_name='PL ASINs')

```
